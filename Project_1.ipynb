```python
## Dropping all missing values
WB_DF6 = WB_DF6.dropna(how = 'any')
WB_DF6.head()
```

```python
## Below I create a dummy variable for V5
WB_DF6['LIT_DUM'] = WB_DF6.V5
mask = WB_DF6.LIT_DUM < 38
column_name = 'LIT_DUM'
WB_DF6.loc[mask, column_name] = 0
WB_DF6.head()
```

```python
mask2 = WB_DF6.LIT_DUM > 37
column_name1 = 'LIT_DUM'
WB_DF6.loc[mask2, column_name1] = 1
WB_DF6.head()
```

```python
## Let's look at LIT_DUM
WB_DF6['LIT_DUM'].value_counts()
```

```python
## I could also recorde it like this.
WB_DF6.loc[(WB_DF6['V5'] > 0) & (WB_DF6['V5'] <= 38), 'LIT_DUM2'] = 'Hi_Lit'
WB_DF6.loc[(WB_DF6['V5'] > 38), 'LIT_DUM2'] = 'Lo_Lit'
WB_DF6.head()
```

```python
## I'll create a dummy for Civil Liberties.
WB_DF6.loc[(WB_DF6['V133'] > 0) & (WB_DF6['V133'] <= 3), 'CIV_DUM'] = 'Hi_Civ'
WB_DF6.loc[(WB_DF6['V133'] > 3), 'CIV_DUM'] = 'Lo_Civ'
WB_DF6.head()
```

```python
## Already we can see some association between LIT_DUM2 and CIV_DUM (i.e., those countries with low lit are also lo civ)
## Let's look at some distribution graphs.
grid = sns.FacetGrid(WB_DF6, row="LIT_DUM2", col="CIV_DUM", margin_titles=True)
grid.map(plt.hist, "V168", bins=np.linspace(0, 100, 20));
```

```python
## We can see above that low civic liberties and low literacy rate are in the lower percentage of female school enrollment.
## Let's look at some factor plots.
with sns.axes_style(style='ticks'):
    g = sns.factorplot("LIT_DUM2", "V168", "CIV_DUM", data=WB_DF6, kind="box")
    g.set_axis_labels("Literacy", "Female School Enrollment");
```
