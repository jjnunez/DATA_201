```python
subpro4 = subpro4.drop(columns='PSU')
subpro4.head()
```

```python
subpro4.columns.values
```

```python

cols = ['SEX4', 'AGE4', 'MARSTAT4', 'RELIGPART4','RELIGEXTE4', 'RELIGDEAL4', 'RELIGPRAY4', 
                  'POMACAT4', 'LOSS4', 'FS4','MMSECAT4', 'PERSKIDSC4', 'DEPDUM']
subpro5 = subpro4.loc[:, cols]
```

```python
## Split data set in independent variables and target variable
IVS = ['SEX4', 'AGE4', 'MARSTAT4', 'RELIGPART4','RELIGEXTE4', 'RELIGDEAL4', 'RELIGPRAY4', 
                  'POMACAT4', 'LOSS4', 'FS4','MMSECAT4', 'PERSKIDSC4']
X = subpro5[IVS] ## Independent variables
y = subpro5.DEPDUM
```

```python
## Below I split the data into training and testing sets
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# import the class
from sklearn.linear_model import LogisticRegression
```

```python
# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
## Above we see the confusion matrix. Since this is a binary classification, we have a 2x2 matrix.
## Above we see 182 and 62 are the actual predictions, and 68 and 37 are incorrect predictions.
## Let's visualize this
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
## Above we see I got a classification rate of 70%, which is not bad accuracy. ((182+62)/(182+62+68+37))
## Precision is about how precise the model is. (62/(37+62))
## Recall is how often the test set it actually predicted by the model.  (62/(68+62))
```

```python
## Let's graphically analyze sensitivity and specificity
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
## With a AUC = .72, given our data, this isn't a bad.
```
